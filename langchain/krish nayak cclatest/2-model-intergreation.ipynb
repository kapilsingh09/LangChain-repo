# 2. Model Integration

This notebook demonstrates how to integrate different language models into your workflow.

```python
import os
from dotenv import load_dotenv
load_dotenv()
os.environ["GOOGLE_API_KEY"] = os.getenv("GOOGLE_API_KEY")
```

```python
from langchain.chat_models import init_chat_model

# Standard Google AI Studio endpoint
model = init_chat_model(
    model="google_genai:gemini-1.5-flash"
)
model
```

```python
res  =model.invoke("Hello how are you?") 
res
```

```python
res.content[-1]
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI

model = ChatGoogleGenerativeAI(model="gemini-1.5-flash")
response = model.invoke("Hello how are you what is your name?")
response.content[-1]
```

```python
model.invoke("write me a 200 words paragraph on artificial intellegence")
```

```python
for chunk in model.stream("while True: print('Hello World')"):
    print(chunk, end="")
```

```python
for chunk in model.stream("why is the sky blue?"):
    print(chunk.text, end="", flush=True)
```

#Batch

```python
res = model.batch([
    'how do parrots have colorful feathers?',
    'how do airplanes fly?',
    'what is the difference between a cat and a dog?'
])

for response in res:
    print(response)
```

```python
res = model.batch([
    'how do parrots have colorful feathers?',
    'how do airplanes fly?',
    'what is the difference between a cat and a dog?'
    config={
        'max_corre': 0.9,
    }
])

for response in res:
    print(response.content[-1].text)
```
